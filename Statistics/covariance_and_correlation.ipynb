### Covariance and Correlation

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import numpy as np

# np.cov()#alternate method for calculation covariance
```

```python
data=sns.load_dataset('healthexp')
```

```python
data.head()
```

```python
#covariance
covariance = data[['Year', 'Spending_USD']].cov()
covariance

```

```python
numeric_data = data.select_dtypes(include=['number'])
covariance_matrix = numeric_data.cov()
```

```python
covariance_matrix
```

```python
numeric_data
```

```python
# Pearson Covariance Method
numeric_data.corr()
```

```python
numeric_data.corr(method='pearson')
```

```python
#Spearman Rank Correlation method
numeric_data.corr(method='spearman')
```

```python
#kendall Correlation method
numeric_data.corr(method='kendall')
```
