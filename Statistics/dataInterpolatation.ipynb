## Data Interpolation

### 1 Linear Data Interpolation Techniques

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
# data interpolation using linear approach
x_new=np.linspace(1,5,10) ##crete new x value
y_interp=np.interp(x_new,x,y)# interpolate y
```

```python
print(y)
```

```python
print(y_interp)
print(x_new)
```

```python
plt.scatter(x_new,y_interp)
```

### Cubic Interpolation with scipy

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
# creating a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
#interpolate the data
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
```

```python
print(x,y)
print(x_new,y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

### PolyNomial Interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
#interpolate dataset using polynomial interpolation techniques
p=np.polyfit(x,y,2)# fit a 2nd degree polynomial to the dataset(bending)
```

```python
x_new=np.linspace(1,5,10)
y_interp=np.polyval(p,x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
