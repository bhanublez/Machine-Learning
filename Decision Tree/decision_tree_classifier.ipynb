## Decision Tree Classifier

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
## Independent and dependent feature
x=df.iloc[:,:-1]
y=dataset.target
```

```python
x,y
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## post pruning
treeclassfier=DecisionTreeClassifier()
treeclassfier
```

```python
treeclassfier.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(14,10))
tree.plot_tree(treeclassfier,filled=True)
```

```python
## Post pruning with max_depth=2
treeclassfier=DecisionTreeClassifier(max_depth=2)
treeclassfier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(5,3))
tree.plot_tree(treeclassfier,filled=True)
```

```python
## prediction
y_pred=treeclassfier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
